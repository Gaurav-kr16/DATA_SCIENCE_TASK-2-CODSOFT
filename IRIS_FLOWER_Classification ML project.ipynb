```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
IRIS['species']=le.fit_transform(IRIS['species'])
IRIS.head()
```

```python
IRIS.head(10)
```

```python
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
X = IRIS.drop(columns=['species'])
Y = IRIS['species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
 #logistic regression 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# model training
model.fit(x_train, y_train)
```

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
KNC = KNeighborsClassifier()
```

```python
KNC.fit(x_train, y_train)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(x_train, y_train)
```

```python
model_accuracy=model.score(x_test,y_test)*100
KNC_accuracy=KNC.score(x_test,y_test)*100
DTC_accuracy=DTC.score(x_test,y_test)*100
```

```python
print(f"Accuracy by using Logistic Regression: {model_accuracy}%")
```

```python
print(f"Accuracy by using K Nearest Neighbors Algorithm: {KNC_accuracy}%")
```

```python
print(f"Accuracy by using Decision Tree Classifier: {DTC_accuracy}%")
```
